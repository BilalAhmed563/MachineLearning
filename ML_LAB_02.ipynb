```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data =pd.read_csv('/content/drive/MyDrive/ML_LAB/auto-mpg.csv')
data.head()
```

```python
data.shape
```

```python
ax=data['mpg'].hist(grid=False, bins=np.linspace(10,50,10))
ax.set_ylabel('miles per gallon')
ax.set_xlabel('Frequency')
plt.show
```

#### UNDERSTANDING :
* Most of the cars have low milage(mpg)

```python
data.corr(numeric_only=True)
```

Possitive Corellation: whn\en one thing goes upt the other thing always goes up.
Negative Corellation: whnen one thing goes up the other thing always goes down.

```python
ax = data['displacement'].hist(grid=False, bins=np.linspace(60, 500, 10), color='red', edgecolor='black')
ax.set_xlabel('Displacement')
ax.set_ylabel('Frequency')
plt.title('Histogram of Displacement')
plt.show()
```

Histogram of Displacement:
This represents how engine displacement (size of the engine) varies among cars.
Larger displacement generally means more powerful engines but lower fuel efficiency
The graph can show whether the dataset contains mostly small or large engines.

```python
ax = data['horsepower'].hist(grid=False, bins=np.linspace(40, 240, 10))
ax.set_xlabel('Horsepower')
ax.set_ylabel('Frequency')
plt.title('Histogram of Horsepower')
plt.show()
```

2. Histogram of Horsepower:
This graph shows how many cars fall into different horsepower ranges.
If most values are on the lower side, it means most cars have less power.
A wider spread indicates the dataset has both low and high-power cars.

```python
ax = data['weight'].hist(grid=False, bins=np.linspace(1500, 5500, 10))
ax.set_xlabel('Weight')
ax.set_ylabel('Frequency')
plt.title('Histogram of Weight')
plt.show()
```

```python
ax = data['acceleration'].hist(grid=False, bins=np.linspace(8, 25, 10))
ax.set_xlabel('Acceleration')
ax.set_ylabel('Frequency')
plt.title('Histogram of Acceleration')
plt.show()
```
